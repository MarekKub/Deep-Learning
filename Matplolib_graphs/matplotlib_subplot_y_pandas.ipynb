# Integración con Pandas

Matplotlib se integra con la librería Pandas, permitiendo dibujar gráficos a partir de los datos de las series y DataFrames de Pandas.

A continuación se declaran Dataframes con valores  de temperatura para distintas ciudades y se representa la información en gráficos.


```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.DataFrame({'Días':['L', 'M', 'X', 'J', 'V', 'S', 'D'],
    'Madrid':[28.5, 30.5, 31, 30, 28, 27.5, 30.5],
    'Barcelona':[24.5, 25.5, 26.5, 25, 26.5, 24.5, 25]})
df.plot("Días", "Madrid",  kind="line") # Gráfica para cada ciudad
df.plot("Días", "Barcelona",  kind="line") # Gráfica para cada ciudad

# Dibujar las gráficas en un mismo área, axes
fig, ax = plt.subplots()
df.plot(x = 'Días', y = 'Madrid', ax = ax)
df.plot(x = 'Días', y = 'Barcelona', ax = ax)
plt.show()

```

2. Se muestra en dos subplots, en uno se muestra la población de cada isla, y en otro su área.

```python

import pandas as pd
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))

plt.subplot(1,2,1)
plt.bar(pd_democracias['Island'], pd_democracias['Population'], color="blue")
plt.title('Poblacion')
plt.xlabel("Isla")
plt.ylabel("Población")
plt.xticks(rotation=45)

plt.subplot(1,2,2)
plt.bar(pd_democracias['Island'], pd_democracias['Area'], color="pink")
plt.title('Área por isla')
plt.xlabel("Isla")
plt.ylabel("Área(km2)")
plt.xticks(rotation=45)
```

3) Se crean tres subplots, representado: arriba a la izquierda, un diagrama de barras en rojo con los gastos de un año, agrupado en meses. En la derecha, un diagrama de barras en azul, con los ingresos en meses Y abajo, un diagrama de líneas con con los ingresos frente a los gastos, a lo largo de los meses.

```python

#Ingresos
meses = ['Enero','Febrero','Marzo','Abril','Mayo','Junio','Julio','Agosto','Septiembre','Octubre','Noviembre','Diciembre']
ingresos = [100,300,150,400,150,250,300,300,320,340,55,90]

#Gastos
gastos = [90,150,100,100,120,150,200,100,20,40,55,80]

plt.figure(figsize=(20,20))

plt.subplot(2,2,1)
plt.bar(meses, gastos, color="red")
plt.title("Gastos")
plt.xlabel("Meses")
plt.ylabel("Gastos")
plt.xticks(rotation=45)

plt.subplot(2,2,2)
plt.bar(meses, ingresos, color="blue")
plt.title("Ingresos")
plt.xlabel("Meses")
plt.ylabel("Ingresos")
plt.xticks(rotation=45)

plt.subplot(2,2,3)
plt.plot(meses, ingresos, color="red", marker="v")
plt.plot(meses, gastos, color="blue", marker="s")
plt.legend(["Ingresos", "Gastos"], loc="upper right")
plt.title("Ingresos VS gastos")
plt.xlabel("Meses")
plt.ylabel("Miles de € (k)")
plt.xticks(rotation=45)
```
